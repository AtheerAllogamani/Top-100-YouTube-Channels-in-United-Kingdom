<h2>Loading the data from a CSV and print the first five rows<h2>

```python
# Define librires
import pandas as pd
# Read in the CSV as a DataFrame
Top100 = pd.read_csv("Top 100 YouTube Channels in United Kingdom/Dataset.csv")
# Print the first five rows of the DataFrame
print(Top100[:5])
```

<h2>Creating a plot with Grades in x and Category in y<h2>

```python
# Define librires
import matplotlib.pyplot as plt

# Create a figure
fig = plt.figure()

# Draw a line plot of release_years and durations
plt.plot(Top100['Grade'], Top100['Category'])

# Create a title
plt.title("Top 100 YouTube Channels in United Kingdom")

# Show the plot
plt.show()
```

<h2>Creating a scatter plot with Grades in x and Category in y<h2>

```python
# Define librires
import matplotlib.pyplot as plt

# Create a figure
fig = plt.figure()

# Draw a line plot of release_years and durations
plt.scatter(Top100['Grade'], Top100['Category'])

# Create a title
plt.title("Top 100 YouTube Channels in United Kingdom")

# Show the plot
plt.show()
```

<h2>Define the colors for data<h2>

```python
# Define an empty list
colors = []

# Iterate over rows of netflix_movies_col_subset
for lab, row in Top100.iterrows():
    if row['Grade'] == 'A':
        colors.append("green")
    elif row['Grade'] == "B-":
        colors.append("red")
    else:
        colors.append("black")
        
# Inspect the first 10 values in your list        
colors[:10]
```

<h2>Using the color defintion for colorful plotting<h2>

```python
# Set the figure style and initalize a new figure
plt.style.use('fivethirtyeight')
fig = plt.figure()

# Create a scatter plot of duration versus release_year
plt.scatter(Top100['Grade'], Top100['Category'], c=colors)

# Create a title and axis labels
plt.title("Top 100 YouTube Channels in United Kingdom")
plt.xlabel("Grade")
plt.ylabel("Category")

# Show the plot
plt.show()
```
